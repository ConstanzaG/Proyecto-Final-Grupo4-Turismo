```python
#Librerías necesarias para cargar dataframes a .db
import pandas as pd
import sqlite3
```

```python
turismo = pd.read_csv('BBDD 4X30.csv', sep=';')
turismo.head()
```

```python
turismo.columns
```

```python
#Creamos archivo .db en el directorio del programa
connection = sqlite3.connect('turismo.db') 
c = connection.cursor()
#Creamos la tabla donde se cargará el dataframe
c.execute('CREATE TABLE IF NOT EXISTS TURISMO_RECEPTIVO(Código_Comuna_origen text, Comuna_origen text, Código_Provincia_origen text, Provincia_Origen text, Código_Región_origen text, Región_origen text, Código_Comuna_destino text,' +
                                                'Comuna_destino text, Destino_turístico text, Código_Provincia_Destino text, Provincia_destino text, Código_Región_destino text, Región_destino text,' + 
                                                'Años number, Enero number, Febrero number,	Marzo number, Abril number, Mayo number, Junio number, Julio number, Agosto number, Septiembre number, Octubre number,' +
                                                'Noviembre number, Diciembre number)')
connection.commit()
#Ejemplo de dataframe creado en pandas en base a una colección del tipo diccionario
data = {'Código_Comuna_origen': turismo['Código Comuna origen'],
        'Comuna_origen': turismo['Comuna de origen'],
        'Código_Provincia_origen': turismo['Código Provincia origen'],
        'Provincia_Origen': turismo['Provincia Origen'],
        'Código_Región_origen': turismo['Código Región origen'],
        'Región_origen': turismo['Región de origen'],
        'Código_Comuna_destino': turismo['Código Comuna destino'],
        'Comuna_destino': turismo['Comuna de destino'],
        'Destino_turístico': turismo['Destino turístico '],
        'Código_Provincia_Destino': turismo['Código Provincia Destino'],
        'Provincia_destino': turismo['Provincia de destino'],
        'Código_Región_destino': turismo['Código Región destino'],
        'Región_destino': turismo['Región de destino'],
        'Años': turismo['Años'],
        'Enero': turismo['Enero'],
        'Febrero': turismo['Febrero'],
        'Marzo': turismo['Marzo'],
        'Abril': turismo['Abril'],
        'Mayo': turismo['Mayo'],
        'Junio': turismo['Junio'],
        'Julio': turismo['Julio'],
        'Agosto': turismo['Agosto'],
        'Septiembre': turismo['Septiembre'],
        'Octubre': turismo['Octubre'],
        'Noviembre': turismo['Noviembre'],
        'Diciembre': turismo['Diciembre'],

}
df = pd.DataFrame(data, columns= ['Código_Comuna_origen', 'Comuna_origen', 'Código_Provincia_origen', 'Provincia_Origen', 'Código_Región_origen', 'Región_origen', 'Código_Comuna_destino', 'Comuna_destino',
                                  'Destino_turístico', 'Código_Provincia_Destino', 'Provincia_destino', 'Código_Región_destino', 'Región_destino', 'Años', 'Enero', 'Febrero',	'Marzo', 'Abril', 'Mayo',
                                  'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'])

df.to_sql('TURISMO_RECEPTIVO', connection, if_exists='replace', index = False)
#Consultamos la tabla creada en la base de datos .db para ver que proceso fue exitoso
df = pd.read_sql('SELECT * FROM TURISMO_RECEPTIVO', connection)
df
```

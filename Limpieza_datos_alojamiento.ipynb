## Limpieza de datos

```
# modificar nombre columna region
df1.rename({'Unnamed: 2': 'región'}, axis=1, inplace=True)
df2.rename({'Unnamed: 1': 'región'}, axis=1, inplace=True)
```

```
# eliminar columnas sin nombres
df1.drop(df1.filter(regex="Unname"),axis=1, inplace=True)
df2.drop(df2.filter(regex="Unname"),axis=1, inplace=True)

# eliminar primera fila
borrar1 = df1[ df1['región'] == 'Total Nacional'].index
df1.drop(borrar1 , inplace=True)
borrar2 = df2[ df2['región'] == 'Total nacional'].index
df2.drop(borrar2 , inplace=True)

# Eliminar filas nulas
#df1 = df1.dropna(how='all')
df1 = df1.dropna()
df2 = df2.dropna()
df1
```

```
# Agrupar valores por año 2014 - 2018

df2014 = df1[['región','enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre', 'Total anual']]
df2014 = df2014.assign(Año=2014)

df2015 = df1[['región','enero.1', 'febrero.1', 'marzo.1', 'abril.1', 'mayo.1', 'junio.1', 'julio.1', 'agosto.1', 'septiembre.1', 'octubre.1', 'noviembre.1', 'diciembre.1', 'Total anual.1']]
df2015 = df2015.set_axis(['región','enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre', 'Total anual'], axis=1)
df2015 = df2015.assign(Año=2015)

df2016 = df1[['región','enero.2', 'febrero.2', 'marzo.2', 'abril.2', 'mayo.2', 'junio.2', 'julio.2', 'agosto.2', 'septiembre.2', 'octubre.2', 'noviembre.2', 'diciembre.2', 'Total anual.2']]
df2016 = df2016.set_axis(['región','enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre', 'Total anual'], axis=1)
df2016 = df2016.assign(Año=2016)

df2017 = df1[['región','enero/Ref', 'febrero/Ref', 'marzo/Ref', 'abril/Ref', 'mayo/Ref', 'junio/Ref', 'julio/P', 'agosto/P', 'septiembre/P', 'octubre/P', 'noviembre/P', 'diciembre/P', 'Total anual.3']]
df2017 = df2017.set_axis(['región','enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre', 'Total anual'], axis=1)
df2017 = df2017.assign(Año=2017)

df2018 = df2[['región','enero.1','febrero.1', 'marzo.1', 'abril.1', 'mayo.1', 'junio.1', 'julio.1', 'agosto.1', 'septiembre.1', 'octubre.1', 'noviembre.1', 'diciembre.1', 'Total anual 2018']]
df2018 = df2018.set_axis(['región','enero', 'febrero', 'marzo', 'abril', 'mayo', 'junio', 'julio', 'agosto', 'septiembre', 'octubre', 'noviembre', 'diciembre', 'Total anual'], axis=1)
df2018 = df2018.assign(Año=2018)
```

```python
#Gráfico 
df = pd.read_csv("BBDD 4X30.csv", sep = ";")
df

df = df.groupby(["Región de destino"], sort=False)[["Diciembre"]].count() 
df
grafico = df.plot(kind='bar', figsize=(10,5), xlabel='regiones', title='visitas por region')
```
